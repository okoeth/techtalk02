# Data Exploration

## Loading Data

```python
import pandas as pd
df = pd.read_csv("activities.csv", index_col=0, parse_dates=True, infer_datetime_format=True)
```

```python
df
```

## Checking Data

```python
df = df.query('distance>0.0').copy()
```

```python
df.info()
```

## Converting Data

```python
df['elapsed_time'] =  pd.to_timedelta(df['elapsed_time'])
df.info()
```

```python
df['average_pace'] = 1000/(df['average_speed'] *60)
df
```

## Understanding Data

```python
import matplotlib.pyplot as plt
%matplotlib inline
df.plot(y='average_speed')
df.plot(y='average_heartrate')
df.plot(y='distance')
```

```python
df["type"] = df.average_speed.map( lambda x: 'run' if x > 2.0 else 'walk')
df
```

```python
runs=df.query('type=="run"').copy()
```

```python
runs.plot(y='average_speed')
runs.plot(y='average_heartrate')
runs.plot(y='distance')
```

```python
runs["fitness"]=runs["average_speed"]/runs["average_heartrate"]
runs = runs.query("fitness< 0.03")
runs.plot(y='fitness')
```

## Enriching Data

```python
from time import sleep
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="techtalk02")
geolocator.reverse("48.21,11.3")
```

```python
rich_df = df.query("start_latitude == start_latitude").head().copy()

def lookup_address(lat, lon):
    print("Lookup {}{}".format(lat,lon))
    sleep(2.0)
    loc = geolocator.reverse(str(lat)+", "+str(lon))
    return loc.raw["address"]

def lookup_country(lat, lon):
    return lookup_address(lat, lon)["country"]

def lookup_postcode(lat, lon):
    return lookup_address(lat, lon)["postcode"]

rich_df['location_country'] = rich_df.apply(
    lambda row: lookup_country(row['start_latitude'], row['start_longitude']), axis=1)
rich_df['location_postcode'] = rich_df.apply(
    lambda row: lookup_postcode(row['start_latitude'], row['start_longitude']), axis=1)
rich_df['location_postcode']
```
